```python
# Assuming you have a DataFrame named 'trainfinal' with your data

# Check if the specified columns exist in the DataFrame
columns_to_include = ['Product id', 'Unit Sales', 'Total Price', 'Base Price']  # Include 'Date' if it exists
existing_columns = [col for col in columns_to_include if col in trainfinal.columns]

# Specify the column name you want to use for the correlation analysis
target_column = 'Unit Sales'  # Replace with your target column name

if existing_columns:
    # Compute the Pearson correlation matrix including the specified columns
    correlation = trainfinal[existing_columns].corr(method='pearson')

    # Get the top 8 columns with the highest correlation to the target column
    top_columns = correlation.nlargest(8, target_column)[target_column].index

    # The 'top_columns' variable now contains the column names with the highest correlation
    print(top_columns)
else:
    print("The specified columns do not exist in the DataFrame.")
```

```python
import numpy as np
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt

# Assuming you have already defined 'columns' and 'trainfinal2' appropriately

data = pd.read_csv("demand_data.csv")
trainfinal2 = pd.DataFrame(data)
columns = ['Product id', 'Unit Sales', 'total_price', 'base_price']
# columns = ['B', 'C', 'J', 'K']
print(trainfinal2)

# columns = ['Product id', 'Unit Sales', 'total_price', 'base_price']  # Adjust column names as needed

# # Verify the column names in your DataFrame
# print(trainfinal2.columns)

# Compute the correlation matrix
correlation_map = np.corrcoef(trainfinal2[columns].values.T)

# Set the font scale
sns.set(font_scale=1.0)

# Create the heatmap
heatmap = sns.heatmap(
    correlation_map,
    cbar=True,
    annot=True,
    square=True,
    fmt='.2f',
    yticklabels=columns,
    xticklabels=columns
)

# Show the heatmap
plt.show()
```
